### Statistical Analysis
This section includes calculations of mean, median, quantiles, and interquartile range for both datasets, along with an interpretation of these statistical measures.


### Measures of Location
In this section, we analyze central tendency units of measurement, such as the mean and median, for both the male and female height datasets. These measures help in understanding the typical values (values of common occurence and central tendencies) in our data.


```python
male_mean = np.mean(maleheights)
male_median = np.median(maleheights)
mmale = np.quantile(maleheights, [0, 0.25, 0.5, 0.75, 1])
maleIQR = np.quantile(maleheights, [0.75]) - np.quantile(maleheights, 0.25)


print("the male arithmetic mean: " + str(male_mean))
print("the male arithmetic median: " + str(male_median))
print("Male Quantiles: " + str(mmale))
print("Male IQR is: " + str(maleIQR))


```

#### from the male mean and median height calculation we can see that the mean and median are more or less equal when rounded to the nearest decimal place proving to show a symmetric distribution, this shows to be a more central measurement of central tendency as it looks to be centred when comparing the histogram and calculated mean

```python
female_mean = np.mean(femheights)
female_median = np.median(femheights)
ffemale = np.quantile(femheights, [0, 0.25, 0.5, 0.75, 1])
femIQR = np.quantile(femheights, [0.75]) - np.quantile(femheights, 0.25)
    

print("the female arithmetic mean: " + str(female_mean))
print("the female arithmetic median: " + str(female_median))
print("Female Quantiles: " + str(ffemale))
print("Female IQR is: " + str(femIQR))
```

#### from the measure of location with regards to female heights, it also seems to be symettrically distributed judging from the similarity in the calculated median and mean when rounded to the nearest decimal

#### From the above results we can see that males have a larger IQR and Arithmetic mean/median then females do thus showing that males on average are taller then females (atleast from the specified data from those who were recorded)

### Measures of Dispersion
In this part of our analysis, we focus on understanding the spread of the data around the mean. This is done through the calculation of standard deviation and interquartile range (IQR) for both male and female height datasets.


#### Standard Deviation and IQR

```python
ss = np.std(maleheights)
print("the average degree of spread from the arithmetic mean calculated is (Standard Deviation): " + str(ss)) 
print("Male IQR is: " + str(maleIQR))
print('---------------------------------------------------------------------------')
sts = np.std(femheights)
print("the average degree of spread from the arithmetic mean calculated is (Standard Deviation): " + str(sts)) 
print("Female IQR is: " + str(femIQR))
```

#### From the analysis males are generally taller than females, according to IQR and Standard Deviation analysis. Male heights are more variable, with a Standard Deviation 0.5 units higher than female heights. Furthermore, the male Interquartile Range outperforms the female IQR by 0.8 units, indicating a wider range in the middle 50% of male height. The greater variety and changeability of male heights, combined with the sensitivity of Standard Deviation to outliers, emphasises the differences in height distributions between genders.

### Correlation Analysis
This section involves computing Pearson’s and Spearman’s correlation coefficients for the standardised data. The analysis aims to uncover any linear or rank-order relationships between different physical measurements.


```python
plt.style.use("seaborn")
df = pd.DataFrame(C [: , [0, 1 , 2, 3 ,4]],
                 columns =[ "height", "weight", "waist circumference", "hip circumference", "BMI"]
                 )
df
sns.pairplot(df)
plt.show()
```

```python
clr = cm.get_cmap("copper")
cols = np.array([ "height", "weight", "waist circumference", "hip circumference", "BMI"])
sns.heatmap(C, xticklabels=cols, yticklabels=cols, annot=True, cmap = clr)
plt.show()
```

```python
clr = cm.get_cmap("copper")
o = [4, 3, 0, 2, 1,]
cols = np.array([ "height", "weight", "waist circumference", "hip circumference", "BMI"])
sns.heatmap(C[np.ix_(o,o)], xticklabels=cols[o], yticklabels=cols[o], annot=True, cmap = clr)
plt.show()
```

### Spearman’s correlation coefficients

```python
#Spearmans rho
```

```python
def plot_corr(x,y):
    r = stats.pearsonr(x,y)[0]
    rho = stats.spearmanr(x,y)[0]
    plt.scatter(x,y, label =f"r    = {r:.3}\nrho =  {rho:.3}")
    plt.legend()
```

```python
x = C[:, 1]
y = C[:, 4]
stats.pearsonr(stats.rankdata(x),stats.rankdata(y))[0]
```

```python
stats.spearmanr(x,y)[0]
plot_corr(x,y)
```

```python
plot_corr(C[:, 1], C[:, 2])
# Weight vs waist circumference
```

```python
C = np.corrcoef(stats.rankdata(spear, axis = 0).T)
C
clr = cm.get_cmap("copper")
cols = np.array([ "height", "weight", "waist circumference", "hip circumference", "BMI"])
sns.heatmap(C, xticklabels=cols, yticklabels=cols, annot=True, cmap = clr)
plt.show()
```
